```python
class GC():
    pass
gc = GC()

gc.num_joints = 25
gc.max_nframe = 4
gc.connect_bone = [
        (3, 2), (2, 20), # head
        (20,4), (4,5), (5,6), (6,7), (7,21), (6,22), # left arm
        (20,8), (8,9), (9,10), (10,11), (11,23), (10,24), # right arm
        (20,1), (1,0), # spine
        (0,12), (12,13), (13,14), (14,15), # left foot
        (0,16), (16,17), (17,18), (18,19) # right foot
    ] # the joint id begin from 0

def build_graph(temp_mode):
    g = dgl.DGLGraph()
    g.add_nodes(gc.num_joints * gc.max_nframe)
    connect_bone = np.array(gc.connect_bone)
    connect_bone = np.concatenate((connect_bone, connect_bone[:, ::-1]), axis=0)
    expand_connect_bone = np.zeros((2*24*gc.max_nframe, 2))
    for eacht in range(gc.max_nframe):
        expand_connect_bone[eacht*48:(eacht+1)*48, :] = connect_bone + eacht * 25
    g.add_edges(expand_connect_bone[:, 0], expand_connect_bone[:, 1])
    # spatial connection
    temporal_connect_forward = np.zeros( ((gc.max_nframe-1)*gc.num_joints, 2) )
    for eachj in range(gc.num_joints):
        apoint = np.array([25*i + eachj for i in range(0,gc.max_nframe-1)])
        bpoint = np.array([25*i + eachj for i in range(1,gc.max_nframe)])
        temporal_connect_forward[eachj*(gc.max_nframe-1):(eachj+1)*(gc.max_nframe-1), :] = \
            np.stack((apoint, bpoint), axis=-1)
    
    temporal_connect_backward = temporal_connect_forward[:, ::-1]
    
    if temp_mode in ('forward', 'both'):
        g.add_edges(temporal_connect_forward[:,0], temporal_connect_forward[:,1])
    if temp_mode in ('backward', 'both'):
        g.add_edges(temporal_connect_backward[:,0], temporal_connect_backward[:,1])
    # temporal connection
    # g.ndata['h'] = torch.zeros(gc.max_nframe * gc.num_joints, gc.base_num_channel * gc.max_person)
    
    return g
```

```python
g = build_graph('both')
```

```python
nw.draw(g.to_networkx(), with_labels=True)

plt.show()
```
